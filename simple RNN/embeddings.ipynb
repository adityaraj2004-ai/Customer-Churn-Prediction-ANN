```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a teacher and I love to inspire and teach students',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of life',
    'your videos are good',
]
```

```python
sent
```

```python
# Define the vocabulary size:
voc_size=10000
```

```python
# In the output each row is representing sentence, and each element is telling the index of the word acoording to vocabulary size. 
one_hot_repr= [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
# word Embedding Represenation
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# using pad sequences to equalize the length of the sentences
sent_length = 12
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen= sent_length)
```

```python
print(embedded_docs)
```

```python
# Feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
